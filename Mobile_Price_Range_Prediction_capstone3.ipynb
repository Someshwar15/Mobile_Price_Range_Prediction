### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

2.   Chi-Squared Test for figure out the relation between 'ram' feature and 'price_range' with p value <= 0.5.



### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Chi-Test for figure out the relation between 'mobile_wt' feature and 'price_range' with p value <= 0.5.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
chi_test = pd.crosstab(df['mobile_wt'], df['price_range'])
from scipy.stats import chi2_contingency
stat, pvalue, dof, expected_R = chi2_contingency(chi_test)
print('pvalue :', pvalue)

if pvalue <= 0.5:
  print('Alternetive Hypothesis passed. mobile_wt and price_range have relationship')
else:
  print('Null Hypothesis passed. mobile_wt and price_range does not have relationship')
````

##### Which statistical test have you done to obtain P-Value?

we use the chi-square test to obtain the p-value between the features 'mobile_wt' and 'price_range'.

##### Why did you choose the specific statistical test?

we use chi-squared test tp determine wether there is significant association between two variables. And to compare the observed values in our data to the expected values that we would see if the null hypothesis is true.
